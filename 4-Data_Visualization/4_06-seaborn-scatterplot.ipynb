## 4.5 - Scatter plot in seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Scatter plot
The purpose of a scatter plot is to show the type of relationship, or correlation, that exists between two sets of data. Using a scatter plot, we can observe how scattered the data is.

![Scatter-Plot-of-Blobs-Dataset-Points-Colored-by-Class.png](https://tutorials.one/wp-content/uploads/2020/05/1588507670_936_Scatter-Plot-of-Blobs-Dataset-with-Three-Classes-and-Points-Colored-by-Class-Value-2.png)

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

#### **`seaborn.scatterplot`**
Draw a scatter plot with possibility of several semantic groupings.

```python
plt.figure(figsize=(12, 6))
# sns.scatterplot(x=df['year'], y=df['price'])
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

Each point in the graph represents the values in each row for the `price` column in the table. From the graph we can make the following conclusion:
- Most of the cars were produeced after 2000;
- There's an outlier in year 1970 (and it can disrupt our future computations);
- There's a positive correlation between `price` and `year`, that is, the newer the car, the higher the price.

#### **`seaborn.regplot`**
Plot data and a linear regression model fit.

```python
# cleaning data by year
df = df[df['year'] != 1970]

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='year', y='price')
plt.show()
```

The straight line on the graph is called the **regression line** and shows the relationship between the values on the `x` and `y` axes.

---

Let's see the correlation between the `price` and the `mileage` (distance traveled).

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['mileage'], y=df['price'])
plt.show()
```

From this graph we can see a negative correlation, that is, as the `mileage` increases, the `price` of the car decreses.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=df['mileage'], y=df['price'])
plt.show()
```

Let's add the 3rd variable to our graph, the type of transmission (gearbox) using the `hue` parameter.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission')
plt.show()
```

What conclusions can we get from the graph:
- Most of the cars are **automatic** transmission;
- **Manual** transmission cars are very cheap;
- **Semi-auto** transmission cars are relatively expensive.

---

We use `style` parameter as 4th value in the graph:

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', style='fuelType', s=150)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(data=df, x='year', y='price', hue='fuelType', s=100)
plt.show()
```

#### **`seaborn.swarmplot`**
Draw a categorical scatterplot with non-overlapping points.

> **Attention!** The more categories in swarmplot, the longer the function will run.

> **This code ran exactly `2m 58s` with colab GPU.**

```python
plt.figure(figsize=(16, 8))
sns.swarmplot(data=df, x='transmission', y='price')
plt.show()
```

> **This code ran exactly `4m 20s` on colab CPU.**

```python
plt.figure(figsize=(16, 8))
sns.swarmplot(data=df, x='transmission', y='price')
plt.show()
```

From the graph we can conclude that:
- **Automatic** and **Semi-auto** cars are almost in the same price category
- **Manual** cars are in the segment of cheap cars (up to $30,000).

Draw a scatterplot where one variable is categorical.
