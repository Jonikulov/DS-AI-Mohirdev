## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizing. **Grafikni faylga saqlab oling**

```python
data[4].set_index(0)
```

```python
# JAVOBNI SHUYERGA YOZING
ann_max = df['TM']
ann_min = df['Tm']

plt.figure(figsize=[16, 8])
plt.plot(
    ann_max,
    color="#fc7a00",
    linewidth=2,
    marker="v",
    label = "average max temperature"
)

plt.plot(
    ann_min,
    linestyle="--",
    linewidth=2,
    marker="o",
    label = "average min temperature"
)
plt.grid()
plt.legend()
plt.title("Annual average max and min temperature for Tashkent")
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.savefig('annual_avg_temp.png')
plt.show()
```

```python
#@title To'g'ri javob
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label="O'rtacha minimum harorat")
plt.plot(maxT, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label="O'rtacha maksimum harorat")
plt.grid()
plt.legend()
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('tempMinMax.png')
plt.show()
```
