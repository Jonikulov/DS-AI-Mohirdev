## 4.10 - Subplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### `matplotlib.pyplot.subplot`

Add an Axes to the current figure or retrieve an existing Axes.

```python
x = np.arange(0, 4 * np.pi, .1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 6))
# plt.suptitle("Graphs of the trigonometric functions")
plt.subplot(2, 1, 1)  # 2 - number of rows, 1 - number of columns, 1-graph
plt.plot(x, sin,
         linestyle="-.",
         linewidth=2,
         color='#8E44AD',
         alpha=0.7,
         marker='o'
)
plt.title("Graph of the sin(x) function")
plt.grid(color='tomato', linestyle='--', alpha=0.5)

plt.subplot(2, 1, 2)  # 2 - number of rows, 1 - number of columns, 2-graph
plt.plot(x, cos,
         linestyle=":",
         linewidth=2,
         color='c',
         alpha=0.7,
         marker='v'
)
plt.title("Graph of the cos(x) function")
plt.grid(color='tomato', linestyle='--', alpha=0.5)


plt.show()
```

### `matplotlib.pyplot.subplots`

Create a figure and a set of subplots.

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year >= 2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

`subplots` function returns 2 objects:

  1. __fig__ : Figure - The top level container for all the plot elements.
  2. **ax** : Axes - *ax* can be either a single *Axes* object or an array of Axes objects if more than one subplot was created.

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# A centered suptitle to the figure.
fig.suptitle("Graph 1")
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x='year', y='price', hue='fuelType', s=50)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

sns.histplot(ax=ax[0, 0], data=df, x='price')
sns.histplot(ax=ax[0, 1], data=df, x='mileage')
sns.scatterplot(ax=ax[1, 0], data=df, x='year', y='price', hue='fuelType', s=50)
sns.scatterplot(ax=ax[1, 1], data=df, x='year', y='mileage', hue='fuelType', s=50)

# Add a centered suptitle to the figure.
fig.suptitle("Graph 2")
# Set a title for the Axes.
ax[0, 0].set_title("Price distibution")
ax[0, 1].set_title("Mileage distribution")
ax[1, 0].set_title("Price vs Year correlation")
ax[1, 1].set_title("Mileage vs Year correlation")
plt.show()
```
