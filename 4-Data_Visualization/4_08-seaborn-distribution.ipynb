## 4.8 - Distribution (taqsimot)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Distribution
A **distribution** in statistics is a function that shows the possible values for a variable and how often they occur. The goal of using **distribution graphs** is to illustrate the distribution of each variable. This means showing the value the variable takes across the individuals in the sample.

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
mask = (df.year>=2010) & (df.transmission!='Other') & (df.fuelType!='Other')
df = df[mask]
df.head()
```

#### `seaborn.histplot`

Plot univariate or bivariate histograms to show distributions of datasets.

```python
plt.figure(figsize=(18, 6))
sns.histplot(x='price', data=df)
plt.show()
```

#### `seaborn.displot`

Figure-level interface for drawing distribution plots onto a FacetGrid.

```python
sns.displot(x='price', data=df, height=5, aspect=3)
plt.show()
```

```python
sns.displot(x='mileage', data=df, height=5, aspect=3)
plt.show()
```

```python
sns.displot(x='mileage', hue='transmission', data=df, multiple='stack', height=5, aspect=3)
plt.show()
```

```python
sns.displot(x='mileage', col='transmission', data=df, height=5, aspect=1)
plt.show()
```

#### Kernel Density Estimation (KDE)
**Kernel Density Estimation** is the application of kernel smoothing for probability density estimation, i.e., a non-parametric method to estimate the probability density function of a random variable based on kernels as weights.

##### `seaborn.displot` **kind='kde'**

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=3)
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, height=5, aspect=3)
plt.show()
```

##### `seaborn.kdeplot`
Plot univariate or bivariate distributions using kernel density estimation.

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(data=df, x='price', shade=True)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(data=df, x='price', hue='transmission', shade=True)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', kind='kde', height=5, aspect=3)
plt.show()
```

```python
sns.displot(data=df, x='price', col='fuelType', kind='kde', height=5, aspect=1)
plt.show()
```

##### two-dimensional kernel density estimation (2D-kde)
Plot a bivariate distribution

```python
df.head()
```

```python
sns.displot(data=df, x='price', y='year', kind='kde', height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='price', y='year')
plt.show()
```

#### Cumulative Distribution Function (CDF)
**Cumulative distribution function** is used to describe the probability distribution of random variables. It can be used to describe the probability for a discrete, continuous or mixed variable.

##### `seaborn.displot` **kind='ecdf'**

```python
sns.displot(data=df, x='price', kind='ecdf', height=5, aspect=2)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', kind='ecdf', height=5, aspect=2)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', kind='ecdf', height=5, aspect=2)
plt.grid()
plt.show()
```

##### `seaborn.ecdfplot`

Plot empirical cumulative distribution functions.

```python
plt.figure(figsize=(10, 5))
sns.ecdfplot(data=df, x='price', hue='transmission')
plt.grid()
plt.show()
```
