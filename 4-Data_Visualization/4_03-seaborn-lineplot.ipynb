## 4.3 - **`seaborn`** library. `seaborn.lineplot`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**Seaborn** is a Python data visualization library based on `matplotlib`. It provides a high-level interface for drawing attractive and informative statistical graphics.
![seaborn-logo.svg](https://seaborn.pydata.org/_static/logo-wide-lightbg.svg)

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.shape
```

```python
plt.plot(df)
plt.show()
```

### **`seaborn.lineplot`**
Draw a line plot with possibility of several semantic groupings.

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Graph of visits to museums")
plt.xlabel("Date")
plt.ylabel("Number of people")
sns.lineplot(data=df)
plt.show()
```

#### `plt.xticks`
Get or set the current tick locations and labels of the x-axis.

```python
plt.figure(figsize=(16,6))
plt.title("Graph of visits to museums")
plt.xlabel("Date")
plt.ylabel("Number of people")
plt.xticks(rotation=90)
sns.lineplot(data=df)
plt.show()
```

**The main function of line graphs is to monitor the change (trend) of a process/event over time.**

From the above graph, we can see the number of visitors to different museums in a beautiful and, most importantly, understandable way, and make various conclusions.

For example, among the museums, the *Avila Adobe* museum is very famous. And the *Firehouse Museum* was overcrowded on 2014-09-01 (this may also indicate an error in the data).

---

Here is the graph for *Avila Adobe* museum itself:

```python
plt.figure(figsize=(16,6))
plt.title("Graph of visits to museums")
plt.xlabel("Date")
plt.ylabel("Number of people")
plt.xticks(rotation=90)
sns.lineplot(data=df['Avila Adobe'])
plt.show()
```

Graph for *Avila Adobe* and *Chinese American Museum*

```python
plt.figure(figsize=(16,6))
plt.title("Graph of visits to museums")
plt.xlabel("Date")
plt.ylabel("Number of people")
plt.xticks(rotation=90)
sns.lineplot(data=df[['Avila Adobe', 'Chinese American Museum']])
plt.show()
```
