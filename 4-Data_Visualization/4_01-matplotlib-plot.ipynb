```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

#### **`plot()` properties**
- `linewidth` or `lw` - Line width, in points.

- `linestyle` or `ls` - style of the line
  - `'-'` or `'solid'` - solid line
  - `'--'` or `'dashed'` - dashed line
  - `'-.'` or `'dashdot'` - dash-dotted line
  - `':'` or `'dotted'` - dotted line
  - `'none'`, `'None'`, `' '`, or `''` - draw nothing

- `color` or `c` - color of the line
  - `'b'` - blue
  - `'c'` - cyan
  - `'g'` - green
  - `'k'` - black
  - `'m'` - magenta
  - `'r'` - red
  - `'y'` - yellow
  - `'#FF5733'` (HEX)
  - `(0.1, 0.2, 0.5, 0.3)` (RGBA)

- `alpha` - line transparency (0-1 range, inclusive)

- `marker` - line marker
  - `'.'` - point
  - `'o'` - circle
  - `'v'` - triangle_down
  - `'^'` - triangle_up
  - `'s'` - square
  - `'*'` - star
  - ...
  - full list -> https://matplotlib.org/3.5.2/api/markers_api

```python
plt.plot(x, y, linewidth=2.0, linestyle='--')
plt.show()
```

```python
plt.plot(x, y,
         linewidth=2,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='s'
         )
plt.show()
```

### **`plt.title`, `plt.xlabel`, `plt.ylabel`, `plt.grid`**

```python
plt.plot(x, y,
         linewidth=1.5,
         linestyle='-.',
         color=(0.1, 0.2, 0.5),
         alpha=0.7,
         marker='H')

# title of a graph
plt.title('The graph of the function sin(x)')

# label for the x-axis
plt.xlabel("pi")

# label for the y-axis
plt.ylabel("sin(x)")

# configure the grid lines
plt.grid()
plt.show()
```

### **`plt.figure`**
Create a new figure, or activate an existing figure.
- **figsize*(float, float)***, (default: `[6.4, 4.8]`) - Width, height in inches.

```python
plt.figure(figsize=(10,4))

plt.plot(x, y,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='H')

# Set a title for the Axes
plt.title('The graph of the function sin(x)')

# Set the label for the x-axis
plt.xlabel("pi")

# Set the label for the y-axis
plt.ylabel("sin(x)")

# Configure the grid lines
plt.grid()
plt.show()
```

drawing 2 graphs

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# graph of sin(x)
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# graph of cos(x)
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# # title of a graph
plt.title('Graphs of the sin(x) and cos(x) functions')
# label for the x-axis
plt.xlabel("pi")
# info about each line
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.show()
```

### **`plt.savefig`**
Save the current figure.

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# graph of sin(x)
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# graph of cos(x)
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# # title of a graph
plt.title('Graphs of the sin(x) and cos(x) functions')
# label for the x-axis
plt.xlabel("pi")
# info about each line
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.savefig('sin-cos-graph.png')
```
