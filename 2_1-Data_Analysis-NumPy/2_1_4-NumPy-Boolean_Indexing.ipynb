## 2.1.4 - Boolean Indexing

```python
import numpy as np
```

```python
# Names
names = np.array(['Hasan', 'Husan', 'Mirzabek', 'Elyor', 'Hasan', 'Javohir', 'Elyor'])
names
```

```python
# Data
data = np.random.randn(7, 4)
data
```

```python
names == "Javohir"
```

```python
data[names == "Javohir"]  # Javohir's data
```

```python
data[names == "Elyor"]  # Elyor's data
```

```python
data[names == "Hasan", 2:]  # Hasan's data's last 2 columns
```

```python
data[names != "Hasan"]  # returns everyone's data except Hasan's
```

```python
data[~(names == "Hasan")] # "~"" equivalent to "!="
```

```python
mask = (names=="Hasan") | (names=="Elyor")  # "|" -> or
```

```python
data[mask]  # Hasan and Elyor's data
```

```python
data[data < 0]
```
