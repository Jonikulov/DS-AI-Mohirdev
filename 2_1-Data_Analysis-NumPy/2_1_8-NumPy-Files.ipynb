## 2.1.8 - Working with files

### `save`, `load`, `savez`, `savez_compressed`

```python
import numpy as np
```

```python
arr = np.arange(10)
arr
```

```python
# save an array to a binary file in NumPy .npy format
np.save("data.npy", arr)
```

```python
# load arrays or pickled objects from .npy, .npz or pickled files
arr1 = np.load("data.npy")
arr1
```

```python
arr2 = np.arange(20)
arr2
```

```python
# save several arrays into a single file in uncompressed
np.savez("arrays.npz", a=arr, b=arr2)
```

```python
arrays = np.load("arrays.npz")
```

```python
arrays['a']
```

```python
arrays['b']
```

```python
# save several arrays into a single file in compressed .npz format
np.savez_compressed("arrays_comp.npz", c=arr, d=arr2)
```

```python
arrays_comp = np.load("arrays_comp.npz")
```

```python
arrays_comp['c']
```

```python
arrays_comp['d']
```

```python
# save an array to a text file
np.savetxt("arr_text.txt", arr2)
```

```python
# load data from a text file
arr_text = np.loadtxt("arr_text.txt")
arr_text
```
