## 2.2.1 - Series

```python
import pandas as pd
from pandas import Series
import numpy as np
```

```python
obj = Series([4, -5, 1.4, 8, -2, -18.06])
obj
```

```python
obj.values  # values
```

```python
obj.index  # indexes
```

```python
print(obj[0])
print(obj[5])
```

```python
obj2 = Series([8, 1, 6, 4, 2], index=['a', 'b', 'c', 'd', 'e'])
obj2
```

```python
obj2.index
```

```python
obj2['d']
```

```python
obj2[['c', 'e', 'd']]
```

```python
obj2[obj2 < 5]
```

```python
obj2[obj2 > 4]
```

```python
obj2 * 2  # multiplying each element in array
```

```python
np.exp(obj2)  # calculating exp for each element
```

```python
# adding new value
obj2['b'] = 50
obj2
```

#### creating Seres object from a dictionary

```python
# Series objects behave like a dictionary
print('a' in obj2)
print('g' in obj2)
```

```python
cars_dict = {'Malibu':40000, 'Lacetti':20000, 'Toyota':45000, 'Mazda':52000, 'Honda':38000}
type(cars_dict)
```

```python
cars1 = Series(cars_dict)
type(cars1)
```

```python
cars1
```

```python
cars1 = Series({'Malibu':40000, 'Lacetti':20000, 'Toyota':45000, 'Mazda':52000, 'Honda':38000})
```

```python
models = ['Honda','Mazda','Lacetti','Jaguar','Malibu','Toyota', 'BMW', 'Mercedes']
# changing indexes and additional indexes
cars2 = Series(cars_dict, index=models)
cars2
```

### Series methods

```python
cars2.isnull()  # detect missing values
```

```python
cars2.notnull()  # detect existing (non-missing) values
```

```python
# adding arrays element-wise 
cars1.add(cars2)
```

```python
cars1 + cars2
```

```python
# naming object
cars2.name = "Autosalon"
```

```python
# naming indexes
cars2.index.name = "Models"
cars2
```

```python
cars2.hasnans  # return True if object has any nans
```

```python
cars2.dtype  # return the dtype object of the underlying data
```

```python
cars2.is_unique  # return boolean if values in the object are unique
```

```python
cars2.shape  # return a tuple of the shape of the underlying data
```

```python
cars2.size # return the number of elements in the underlying data
```

```python
cars2.loc['Mazda']  # access a group of rows and columns by label(s) or a boolean array
```

```python
cars2.loc[['Mazda', 'BMW', 'Honda']]
```

```python
cars2.iloc[2]  # purely integer-location based indexing for selection by position
```

```python
cars2.iloc[[2, 1, 4]]
```

```python
cars2.min()  # return the minimum value
```

```python
cars2.max()  # return the maximum value
```

```python
cars2.mean()  # return the mean of the values over the requested axis
```
