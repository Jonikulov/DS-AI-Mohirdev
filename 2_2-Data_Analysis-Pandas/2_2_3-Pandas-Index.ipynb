### `Index` object may have repeatable values

```python
usernames = {"anvar":"anvarbek223", "olim":"olimtoy", "maryam":"mary334455","javohir":"javohir445"}
emails = {"anvar":"anvar@gmail.com", "olim":"olim@mail.ru", "maryam":"mary33@gmail.com","javohir":"javohir445@gmail.com"}

data = {
    'username':usernames,
    'email':emails
    }
```

```python
df = pd.DataFrame(data)
df
```

```python
data2 = {
    'username':{"olim":"olimjon21", "maryam":"maryam2020"},
    'email':{"olim":"olimjon21@mail.ru", "maryam":"maryam20@gmail.com"}
    }
```

```python
df2 = df.append(pd.DataFrame(data2))
df2
```

```python
df2.index
```

### `Index` methods and parameters

```python
newind = pd.Index(['olim','nafisa','hamida'])
newind
```

#### return a boolean array where the index values are in other index values

```python
newind.isin(df2.index)
```

```python
df2.index.isin(newind)
```

#### return a new `Index` with elements of index not in other index values

```python
newind.difference(df2.index)
```

```python
df2.index.difference(newind)
```

#### form the union of two `Index` objects

```python
df2.index.union(newind)
```

#### `delete` and `drop` methods

```python
newind
```

```python
# make new Index with passed location(-s) deleted
newind.delete(2)
```

```python
df2.index
```

```python
# make new Index with passed list of labels deleted
df2.index.drop('olim')
```

#### `insert` method

```python
# make new Index inserting new item at location
df2.index.insert(3, 'jasur')
```

#### `is_monotonic` and `is_monotonic_increasing` attribute

```python
df2.index.is_monotonic  # alias for is_monotonic_increasing
```
