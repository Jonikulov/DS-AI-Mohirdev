### Split data into *train set* and *test set*

**`sklearn.model_selection.train_test_split`** - Split arrays or matrices into random train and test subsets.
    
- `random_state` - Controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls.

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_set
```

```python
test_set.shape
```

```python
df['median_income'].hist(bins=50, figsize=[7, 4])
plt.show()
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins=[0., 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1,2,3,4,5])
df['income_cat'].hist(figsize=[7, 4])
plt.show()
```

```python
df.head()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

stratified_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
# stratified_split.split function returns indices
for train_idx, test_idx in stratified_split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_idx]
    strat_test_set = df.loc[test_idx]
```

```python
strat_train_set['income_cat'].hist(figsize=[7, 4])
plt.show()
```

Balance saved.

```python
strat_train_set.drop('income_cat', axis=1, inplace=True)
strat_test_set.drop('income_cat', axis=1, inplace=True)
```

### Create New Parameters

We saw that `median_incom` is a good parameter to find median price of houses, but this parameter alone is not enough. In shuch cases, we can create new parameters by combining other parameters.

For instance, the total number of rooms in an area (`total_rooms`) is not very important, but dividing the total number of rooms by the number of `households` gives the average number of rooms per household (average number of rooms per house) which can be important in determining the price of the house.

```python
housing['rooms_per_household'] = housing['total_rooms']/housing['households']
housing['bedrooms_per_room'] = housing['total_bedrooms']/housing['total_rooms']
housing['population_per_household'] = housing['population']/housing['households']
```
