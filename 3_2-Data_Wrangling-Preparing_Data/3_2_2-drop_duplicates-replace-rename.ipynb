## 3.2.2 - Drop duplicates, replace, rename

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv')
df.head()
```

### `DataFrame.duplicated`
Return boolean Series denoting duplicate rows.

```python
df.duplicated()
```

```python
df.shape
```

### `DataFrame.drop_duplicates`
Return DataFrame with duplicate rows removed.

```python
df.drop_duplicates().shape
```

#### `subset` parameter
`subset` - column label or sequence of labels, optional. *Only consider certain columns for identifying duplicates*, by
default use all of the columns.

```python
df.drop_duplicates(subset=['country'])
```

```python
df.drop_duplicates(subset=['country', 'rank'])
```

### `Series.map`
`map()` function from Series is used to substitute each value in a Series with another value

```python
df1 = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv",
                  index_col=0)
df1.head()
```

In the `num-of-cylinders` column, the values ​​are written in letters, not numbers. Let's change these values ​​to numbers.

First, let's look at all the available values ​​in this column:

```python
df1['num-of-cylinders'].unique()
```

Create a dictionary object for the values ​​to be replaced:

```python
dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12}
```

Using `.map()` let's add values in `dict` ​​to `df` in the form of a new column called `cylinders`:

```python
df1['cylinders'] = df1['num-of-cylinders'].map(dict)
df1.head()
```

Other functions can be applied to column values ​​using `.map ()`. For example, let's find the value of car prices in *soums*:

```python
soums = 10720 # 1$ = 10720 UZS
df1['price'].map(lambda x: x*soums)
```

```python
# let's add it to the DF
df1['price_uzs'] = df1['price'].map(lambda x: x*soums)
df1.head()
```

### `DataFrame.replace`
Replace values given in `to_replace` with value.

```python
# numeric, str or regex
df1.replace('four', 4)
```

```python
df1.replace(to_replace='five', value=5)
```

```python
# list of str, regex, or numeric
df1.replace(['four', 'six'], [4, 6])
```

```python
# dictionary
df1.replace({'four':4, 'five':5, 'six':6})
```

### `DataFrame.rename`
Alter axes labels.

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv',
                  index_col='country')
df2.head()
```

```python
df2.rename(index=str.upper, columns=str.upper)
```

```python
df2.rename(index={'Macau':'MAKAO'},
           columns={'pop2021':'Population'})
```
