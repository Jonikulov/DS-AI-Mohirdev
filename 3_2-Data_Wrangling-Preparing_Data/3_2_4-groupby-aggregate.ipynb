## 3.2.4 - groupby

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv",
                 index_col=0)
df.head()
```

### `DataFrame.groupby`
Group DataFrame using a mapper or by a Series of columns.

```python
df.groupby(by="company").max()
```

We can also get values ​​for a single column:

```python
df.groupby("company").price.mean()
```

We can specify more than one column in grouping:

```python
df.groupby(['company', 'body-style']).min()
```

### `pandas.DataFrame.agg` or `pandas.DataFrame.aggregate`
Aggregate using one or more operations over the specified axis.

*agg* is an alias for *aggregate*. Use the alias.

```python
df.groupby('company').agg(func=[len, min])
```

```python
df.groupby('body-style').price.agg([max, np.mean])
```
