## 3.2.5 - Outliers. Random Sample

```python
import pandas as pd
import numpy as np
```

### Detecting and Handling Outliers

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/MOCK_USER_DATA.csv",
                 index_col=0)
df.head()
```

```python
df.describe()
```

#### working on **age** column

filter data which has age greater than 123

```python
df[df['age'] > 123]
```

filter data which has age less than 0

```python
df[df['age'] < 0]
```

```python
# combine filters
df[(df['age'] > 123) | (df['age'] < 0)]
```

let's turn negative age values into positive with `numpy.abs`

```python
df['age'] = df['age'].apply(np.abs)
df.head()
```

```python
# check if there is no any negative value
df[df['age'] < 0]
```

let's drop age values that are greater than 123

```python
# get index of rows to be droppped
idx = df[df['age'] > 123].index
# drop outliers
df.drop(index=idx, inplace=True)
# describe dataframe
df.describe()
```

#### working on **gender** column

```python
df['gender'].value_counts()
```

filtering only **Male**s and **Female**s

```python
df[df['gender'].isin(['Male', 'Female'])]
```

### Shuffle DataFrame Rows

#### `numpy.random.permuatation`

Randomly permute a sequence, or return a permuted range.

```python
# example
np.random.permutation(20)
```

#### `DataFrame.take`
Return the elements in the given *positional* indices along an axis.

```python
# example 1
df.take([0,1,2,3,6,10,30,80])
```

```python
# example 2
df.take(np.arange(10))
```

### Random Sampling

#### `DataFrame.sample`
Return a random sample of items from an axis of object.

```python
df.sample(3)
```

##### `replace` parameter
**replace :** ***bool, default False*** -> Allow or disallow sampling of the same row more than once.


```python
df.sample(10, replace=True)
```

Note that the `62` index came out twice.
