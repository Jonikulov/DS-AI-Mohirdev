### Computations / descriptive stats

`count` - Count non-NA cells for each column or row

`cummax`- cumulative maximum over a DataFrame or Series axis

`cummin` - cumulative minimum over a DataFrame or Series axis

`cumprod` - cumulative product over a DataFrame or Series axis

`cumsum` - cumulative sum over a DataFrame or Series axis

`diff` - first discrete difference of element

`kurt` - unbiased kurtosis over requested axis

`mad` - mean absolute deviation of the values over the requested axis

`median` - the median of the values over the requested axis

`pct_change` - percentage change between the current and a prior element

`prod` - product of the values over the requested axis

`var` - unbiased variance over requested axis (dispersiya)

`std` - sample standard deviation over requested axis (standart chetlanish)

`skew` - unbiased skew over requested axis
