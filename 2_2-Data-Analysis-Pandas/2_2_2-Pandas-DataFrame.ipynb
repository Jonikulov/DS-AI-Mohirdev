## 2.2.2 - DataFrame

```python
import pandas as pd
```

#### creating DataFrame from a dictionary

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
        'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
        'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]
        }
```

```python
df = pd.DataFrame(data)
df
```

```python
# return the first n rows. n:int, default 5
df.head()
```

```python
df.head(3)
```

```python
# assigning order of columns
df = pd.DataFrame(data, columns=['Yil','Temp','Aholi soni'])
df.head()
```

```python
df = pd.DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni', 'YaIM'])
df.head()
```

#### like `Series` we can give our own values to indexes in `DataFrame` with **`index`** parameter

```python
df = pd.DataFrame(data, columns=['Yil','Temp','Aholi soni','YaIM'], index=[21, 20, 19, 18, 17, 16, 15, 10])
df
```

#### we can extract any column of `DataFrame` as `Series` object

##### as a attribute

```python
obj = df.Yil
obj
```

```python
type(obj)
```

##### as a dictionary key

```python
obj2 = df['Aholi soni']
obj2
```

```python
type(obj2)
```

```python
# giving values to any column
df.YaIM = 1600
df
```

#### with `Series` we can change any values in columns, in this case we have to define index of each row

```python
val = pd.Series([1750, 1560, 1270], index=[21, 18, 10])
```

```python
df.YaIM = val
df
```

#### in order to change values in columns we can define list which has equal length to rows of the column

```python
YAIM = [1700, 1670, 1600, 1590, 1400, 1300, 1350, 1360]
```

```python
df.YaIM = YAIM
df
```

#### in this example, we are creating a new column called "1500+", then values of this column will be True if values of "YaIM" column is greater than 1500 and vice versa

```python
df["1500+"] = df.YaIM > 1500
df
```

#### like a `dictionary` we can remove columns from `DataFrame` with **`del`** operator

```python
del df["1500+"]
df
```

#### creating `DataFrame` with nested dictionaries

```python
data2 = {
    'Uzbekistan':{2021:33.9, 2020:33.5, 2019:32.9, 2018:32.5, 2017:31.9},
    'Kazakhstan':{2021:18.9, 2020:18.7, 2019:18.5, 2018:18.3, 2017:18.1, 2016:17.8}
}
```

```python
df2 = pd.DataFrame(data2)
df2
```

#### we can give a name to `DateFrame` index

```python
df2.index.name = "Yil"
df2
```
