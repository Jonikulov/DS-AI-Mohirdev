## 2.2.4 - Index, Select, Slice, Filter

```python
import pandas as pd
import numpy as np
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
       'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]
        }
```

### `loc`/`iloc` and `at`/`iat`

#### `.loc` - access a group of rows and columns by label(s) or a boolean array

```python
df1.loc[[2020]]
```

```python
df1.loc[[2018, 2015]]
```

```python
df1.loc[[2020, 2018, 2016], ["Uzbekistan", "Kazakhstan"]]
```

#### `.iloc` - purely integer-location based indexing for selection by position

```python
df1.iloc[[0, 2, 4]]
```

```python
df1.iloc[[1, 3], [0, 1, 2]]
```

#### `.at` - access a single value for a row/column label pair

```python
df1
```

```python
df1.at[2018, "Uzbekistan"]
```

```python
df1.at[2020, "Uzbekistan"] - df1.at[2015, "Uzbekistan"]
```

#### `.iat` - access a single value for a row/column pair by integer position

```python
df1.iat[2, 1]  # row=2, column=1
```

```python
obj = pd.Series(np.arange(10.0, 21.0))
obj
```

```python
obj[0]
```

```python
obj[-1]  # causes an error!
```

This method does not work if indexes containted from integers. What if indexes are not integer but string?

```python
ser = pd.Series(np.arange(5.), index=['a', 'b', 'c', 'd', 'e'])
ser
```

```python
ser[0]
```

```python
ser[-1]  # it works fine!
```

Solution: using `loc`/`iloc` attributes

```python
obj.loc[3]
```

```python
obj.iloc[-1]
```

```python
ser.loc['d']
```

```python
ser.iloc[-1]
```
