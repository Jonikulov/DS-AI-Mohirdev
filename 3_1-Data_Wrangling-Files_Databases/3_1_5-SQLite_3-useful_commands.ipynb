### Useful commands

#### DISTINCT
`SELECT DISTINCT` clause to remove duplicate rows in the result set

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query, db)
countries
```

extract data about Uzbekistan

```python
'Uzbekistan' in countries.country.values
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan'"
dfuz = pd.read_sql(query, db)
dfuz.head()
```

```python
query = """
SELECT country, year, coal, gas, hydro, oil
FROM elect
WHERE country='Uzbekistan' AND year > 2010
"""
dfuz = pd.read_sql(query, db)
dfuz
```

#### `BETWEEN`
`BETWEEN` operator is a logical operator that tests whether a value is in range of values

```python
query = """SELECT *
FROM elect
WHERE country='Uzbekistan' AND year BETWEEN 2010 AND 2021
"""
pd.read_sql(query, db)
```

#### `IN`
`IN` operator determines whether a value matches any value in a list or a subquery

```python
query = """
SELECT *
FROM elect
WHERE country='Uzbekistan' AND year IN (1990,2000,2005,2010,2015,2020);
"""
pd.read_sql(query, db)
```

#### `LIKE`

We use `LIKE` operator to query data based on pattern matching.

SQLite provides two wildcards for constructing patterns. They are percent sign `%` and underscore **`_`** :

- The percent sign `%`** wildcard matches any sequence of zero or more characters.

- The underscore **`_`** wildcard matches any single character.

---
filter data which country names ends with '...stan'

```python
query = """
SELECT *
FROM elect
WHERE country LIKE '%stan' AND year BETWEEN 2019 AND 2021
"""
pd.read_sql(query, db)
```

filter data which country codes have 3 letters and starts with 'A'

```python
query = """
SELECT *
FROM elect
WHERE code LIKE 'A__' and year==2020
"""
pd.read_sql(query, db)
```
