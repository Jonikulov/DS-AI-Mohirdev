## 3.1.4 - SQLite 2 - Write to SQL

```python
import pandas as pd
import sqlite3 as sql
```

### Load DataFrame to SQL Table

creat a new database

```python
db = sql.connect("newdb.db")
cursor = db.cursor()
```

download csv as df

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.head()
```

let's modify column names before loading df to sql table in order to avoid writing "*painful*" sql queries )

```python
cols = ['country','code','year','coal','gas','hydro','renewable_other','solar','oil','wind','nuclear']
df.columns = cols
df.head()
```

**`DataFrame.to_sql`** - write records stored in a DataFrame to a SQL database

```python
# load df to sql table
df.to_sql("elec", db)
```

```python
# list all table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall()
```

```python
# table info
cursor.execute("PRAGMA table_info(elec)")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elec", db, index_col='index')
```

### Create a new table in SQL Database

`db.commit` - save (commit) the changes

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
  rank INT,
  country TEXT PRIMARY KEY,
  density DOUBLE,
  population DOUBLE NOT NULL,
  area INT NOT NULL
);
"""

cursor.execute(query)
db.commit()
```

```python
# list all table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall()
```

```python
# table info
cursor.execute("PRAGMA table_info(worldpop)")
cursor.fetchall()
```

```python
# read (empty) table as df
pd.read_sql("SELECT * FROM worldpop", db)
```

#### insert data into a sql table

```python
query = """
INSERT INTO worldpop VALUES (1, 'Macau', 21946.4667, 658.3940, 30)
"""

cursor.execute(query)
db.commit()
```

```python
pd.read_sql("SELECT * FROM worldpop", db)
```

```python
query = """
INSERT INTO worldpop (country, population, area)
VALUES ('Uzbekistan', 33935.7630, 447400)
"""

cursor.execute(query)
db.commit()
```

```python
# missing values will be None
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

#### update data in the sql table

```python
query = """
UPDATE worldpop
SET rank=133, density=75.8511
WHERE country='Uzbekistan'
"""

cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

#### delete data from sql table

```python
query = "DELETE FROM worldpop WHERE country='Macau'"
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

#### delete a sql table

```python
cursor.execute("DROP TABLE worldpop")
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall()
```

### Writing SQL queries with f-string

load a new dataframe

```python
df2 = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/world_population.csv", index_col='rank')
df2.head()
```

create a new sql table

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
  rank INT,
  country TEXT PRIMARY KEY,
  density DOUBLE,
  population DOUBLE NOT NULL,
  area INT NOT NULL
);
"""

cursor.execute(query)
db.commit()
```

extract certain data from df

```python
uzb = df2[df2.country == 'Uzbekistan']
uzb
```

```python
rank = uzb.index.values[0]
country = uzb.at[rank, 'country']
density = uzb.at[rank, 'density']
population = uzb.at[rank, 'pop2021']
area = uzb.at[rank, 'area']

print(f"rank={rank}, country={country}, density={density}, population={population}, area={area}")
```

```python
query = f"""
INSERT INTO worldpop VALUES ({rank}, '{country}', {density}, {population}, {area})
"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

### close the database
