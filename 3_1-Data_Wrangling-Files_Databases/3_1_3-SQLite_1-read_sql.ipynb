### **SQL** - *Structured Query Language*

**Some of The Most Important SQL Commands**

- `ALTER DATABASE` - modifies a database
- `CREATE TABLE` - creates a new table
- `ALTER TABLE` - modifies a table
- `DROP TABLE` - deletes a table
- `CREATE INDEX` - creates an index (search key)
- `DROP INDEX` - deletes an index

### SQLite

importing modules

```python
import pandas as pd
import sqlite3 as sql
from pprint import pprint as print
```

downloading data

```python
url = "https://www.sqlitetutorial.net/wp-content/uploads/2018/03/chinook.zip"

!wget $url
!unzip "chinook.zip"
```

```python
db = sql.connect("chinook.db")
cursor = db.cursor()
```

#### tables in the *chinook* database

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

#### the number of rows in the table

```python
cursor.execute("SELECT COUNT (*) FROM customers")
cursor.fetchone()
```

#### columns in the table
`PRAGMA` - used to query information about a specific table

```python
# PRAGMA command is specific to SQLite
cursor.execute("PRAGMA table_info(customers);")
cursor.fetchall()
```

#### Read a SQL table into a DataFrame
read only limited rows

```python
query = "SELECT * FROM customers LIMIT 5"
df = pd.read_sql(query, db, index_col="CustomerId")
df
```

read entire table

```python
query = "SELECT * FROM customers"
df = pd.read_sql(query, db, index_col="CustomerId")
df.sample(5)
```

selecting specific columns from table

```python
query = "SELECT FirstName, LastName, City, State, Phone, Email FROM customers"
pd.read_sql(query, db).head(10)
```

```python
query = "SELECT * FROM customers ORDER BY Country"
pd.read_sql(query, db, index_col="CustomerId").head(10)
```

```python
query = "SELECT * FROM customers ORDER BY City DESC"
pd.read_sql(query, db, index_col="CustomerId").head(10)
```

#### `DISTINCT` clause allows us to remove the duplicate rows in the result set

```python
cursor.execute("SELECT DISTINCT Country FROM customers")
cursor.fetchall()
```

```python
# getting result as DF
pd.read_sql("SELECT DISTINCT Country FROM customers", db).head()
```

#### `WHERE` clause is used to filter records.
It is used to extract only those records that fulfill a specified condition.

```python
# filter by country: Brazil
query = "SELECT * FROM customers WHERE Country='Brazil'"
pd.read_sql(query, db, index_col="CustomerId").head(10)
```

```python
# filter by countreis: Brazil and USA

# 1. let's count customers
query = "SELECT COUNT (*) FROM customers WHERE Country IN ('Brazil', 'USA')"
cursor.execute(query)
cursor.fetchone()
```

```python
# 2. read into DF
query = "SELECT Firstname, LastName, Country, Phone FROM customers WHERE Country IN ('Brazil', 'USA')"
pd.read_sql(query, db)
```
