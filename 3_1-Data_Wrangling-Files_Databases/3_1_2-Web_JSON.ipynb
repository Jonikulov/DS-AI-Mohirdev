```python
import pandas as pd
import numpy as np
import requests
import json
```

### `read_html`
Read HTML tables into a list of `DataFrame` objects.

```python
df_list = pd.read_html("https://en.wikipedia.org/wiki/Tashkent")
```

### `read_json`
Convert a JSON string to pandas object.

```python
# library api
r = requests.get("https://openlibrary.org/subjects/data_science.json")
json_data = json.dumps(r.json()['works'])

pd.read_json(json_data)
```

```python
# universities api
r = requests.get("http://universities.hipolabs.com/search?country=uzbekistan")
pd.DataFrame(r.json())
```
